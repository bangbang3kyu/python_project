```python


class StudentScores:
    
    # 생성자 : 파일을 읽고 데이터를 딕셔너리에 저장
    def __init__(self, file_path="./scores_korean.txt"):
        self.file_path = file_path
        self.scores = {}
        self.average = 0.0
        
        try:
            with open(self.file_path, 'r', encoding="utf-8") as f:
                for line in f:
                    # 쉼표를 기준으로 이름과 점수를 분리
                    name, score = line.split(',')
                    # score의 type이 str이므로 int()로 변환 후 딕셔너리에 저장
                    self.scores[name] = int(score)

        except Exception as e:
            print(f"[오류] 파일을 읽는 중 문제가 발생했습니다: {e}")
    
    # 평균 계산
    def calculate_average(self):
        result = 0
        for i in self.scores.values():
            result += i
        self.average = result / len(self.scores)

    # 평균 이상 이름 리스트 반환
    def get_above_average(self):
        above_list = []  # 이름을 담을 빈 리스트 생성
        for name, score in self.scores.items():
            if score >= self.average:
                above_list.append(name)
        
        return above_list
    
    # 평균 미만 학생 저장
    def save_below_average(self):
        # 원본 포맷에 맞춰 쉼표 뒤 공백을 제거했습니다.
        with open("below_average_korean.txt", "w", encoding="utf-8") as f:
            for name, score in self.scores.items():
                if score < self.average:
                    f.write(f"{name},{score}\n")
        
        print("[완료] 평균 미만 학생 명단이 저장되었습니다.")
    
    # 결과 출력
    def print_summary(self):
        print("\n" + "=" * 30)
        print("      [ 성적 분석 결과 ]")
        print("=" * 30)
        
        # 1. 평균 점수 출력
        print(f" 전체 평균 점수: {self.average:.2f}점")
        
        # 2. 앞서 만든 함수를 호출해서 평균 이상 학생 리스트 가져오기
        above_list = self.get_above_average()
        
        # 3. 리스트에 담긴 학생 이름들을 출력
        print(f" 평균 이상 학생: {', '.join(above_list)}")
        print("=" * 30)


# ==========================================
#  실질적으로 코드를 동작시키는 실행부
# ==========================================
if __name__ == "__main__":
    # 1. 인스턴스 생성 (파일 자동 로드)
    analyzer = StudentScores("./scores_korean.txt")
    
    # 데이터가 정상적으로 로드된 경우에만 순서대로 실행
    if analyzer.scores:
        # 2. 평균 계산 수행
        analyzer.calculate_average()
        
        # 3. 결과 화면 출력 (평균 및 이상자 명단)
        analyzer.print_summary()
        
        # 4. 미만자 파일 별도 저장 (`below_average_korean.txt`)
        analyzer.save_below_average()
```
