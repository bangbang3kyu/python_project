```python
korean = int(input("국어 점수를 입력하세요 : "))
english = int(input("영어 점수를 입력하세요 : "))
math = int(input("수학 점수를 입력하세요 : "))
print("국어 : {}점\n영어 : {}점\n수학 : {}점\n총점 : {}점\n평균 : {:.1f}점".format(korean, english, math, korean + english + math, (korean + english + math) / 3))
```

```python
gender = input("성별을 입력하세요 (남성-M/여성-F) : ")
age = int(input("나이를 입력하세요 : "))

if (gender == "남성" or gender == "M") and age >= 18:
    print("성별이 남성이며 18세 이상의 성인입니다.")
else:
    print("조건에 해당하지 않습니다.")
```
