### 클래스의 상속

```python
# 사칙연산을 하는 클래스 
# => 클래스 이름, 변수, 함수(메서드), 초기화 함수(생성자)
# 클래스의 이름은 대문자로 시작 => FourCal

class FourCal:
    # 클래스의 멤버 : 변수, 메서드

    # 초기화 함수 => 변수 초기화, 최초 데이터 생성
    def __init__(self, num1,num2):
        # 인스턴스 변수 생성과 데이터 할당
        self.num1 = num1
        self.num2 = num2

    # 메서드를 통해 변수(인스턴스 변수)를 생성 => 클래스를 통해서 객체로 만들면 변수가
    # 함수 생성시에 self 사용하는 이유 => 인스턴스 멤버를 사용하기 위함!
    
    def printData(self):
        print(f'num1 : {self.num1}, num2 : {self.num2}')

    def add(self):
        return self.num1 + self.num2

    def mul(self):
        return self.num1 * self.num2

    def div(self):
        try:
            return self.num1 / self.num2
        except ZeroDivisionError:
            print("에러: 0으로 나눌 수 없습니다.")
            return 0  # 혹은 None 대신 다른 기본값을 반환하도록 설정 가능

    def sub(self):
        return self.num1 - self.num2

```

```python
# 클래스의 상속 : 새로운 클래스 만드는 것
# 상속 => 기존 클래스의 멤버들을 받아서 내것으로 사용
# 오버라이딩 => 메서드를 수정해서 사용, 처리블록을 수정해서 사용
# MoreFourCal : 새로운 클래스 정의 => pow 제곱을 구해주는 함수를 추가, div 메서드 수정(ZeroDisisionError)
class MoreFourCal(FourCal):
    def pow(self):
        return self.num1**self.num2
    
    # 오버라이딩 : 함수의 이름과 매개변수는 그대로 유지하고 -> 처리부분만 수정해서 정의 
    def div(self):
        return super().div()
```

```python
#cal1 = MoreFourCal() # 인자 값 에러 !! => 생성자 호출
cal1 = MoreFourCal(3,0)
```

```python
cal1.div()
```
