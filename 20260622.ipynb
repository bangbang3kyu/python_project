```python
# 문자열 포맷팅
# 3.6 버전 이후 => f''
user_name = '홍길동'
user_age = 20
```

```python
# 안녕하세요! 저는 000 입니다. 나이는 00살 입니다.
print('안녕하세요! 저는 '+user_name+' 입니다. 나이는 '+str(user_age)+'살 입니다')
print(f'안녕하세요! 저는 {user_name} 입니다. 나이는 {user_age}살 입니다.')
print('안녕하세요! 저는 {} 입니다. 나이는 {}살 입니다.'.format(user_name, user_age))
```
