```python
# 숫자를 입력 받아 짝수, 홀수를 판별하는 프로그램
num = int(input("숫자를 입력하세요."))

if num%2==0 and num !=0:
    print("짝수입니다.")
elif num%2==1:
    print("홀수입니다.")
```

```python
# 숫자를 입력 받아 양수, 음수, 0을 판별하는 프로그램
num2 = int(input("숫자를 입력하세요."))

if num2 > 0:
    print("양수입니다.")
elif num2 < 0:
    print("음수입니다.")
else:
    print("0입니다.")
```

```python
# 독감예방 접종이 가능한지 여부를 확인하는 프로그램
age = int(input("나이를 입력하세요."))

if age < 15 or age>=65:
    print("무료예방접종이 가능합니다.")
else:
    print("돈 내셔야합니다.")
```
