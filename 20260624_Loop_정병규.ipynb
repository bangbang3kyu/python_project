```python
# 1부터 100까지 합을 구하는 프로그램 작성
result = 0

for i in range(1,101):
    result+=i

print(f"1부터 100까지의 합은 {result}이다")
```

```python
# for 문을 이용하여 1부터 10까지를 곱한 결과를 출력하는 프로그램 작성
result2 = 1

for i in range(1,11):
    result2*=i
    print(result2)

print(f"1부터 10까지 곱한 결과는 {result2}이다")
```

```python
# 구구단의 짝수 단(2,4,6,8)만 출력하는 프로그램 작성
for i in range(2, 10, 2):
    for j in range(1,10):
        if i >= j:
            print(f"{i}X{j} = {i*j}")
```

```python
num = 0       # 합계를 저장할 변수 (0부터 시작)
c = 1         # 1부터 시작할 숫자 변수
count = 0     # 실제 숫자가 더해진 횟수를 세고 싶다면 사용

while True:
    # c가 짝수(2의 배수)이거나 3의 배수일 때만 더하기
    if c % 2 == 0 or c % 3 == 0:
        num += c
        count += 1
        # (선택 사항) 더해질 때마다 과정을 보고 싶다면 주석을 해제하세요.
        # print(f"더한 수: {c}, 현재 합계: {num}, 총 {count}번 더함")
    
    # 합계가 1000 이상이 되면 무한 루프 탈출
    if num >= 1000:
        break
        
    c += 1    # 다음 숫자로 넘어감

print(f"=== 결과 ===")
print(f"합이 1000을 넘어서는 순간의 마지막에 더한 수(언제): {c}")
print(f"1000이 넘어선 최종 값은: {num}")
```
