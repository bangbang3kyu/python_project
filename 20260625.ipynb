```python
# 고차함수 => 매개변수로 함수를 받아 사용하는 것
# 함수도 객체다!
# filter => 편별 함수(논리값)가 필요
def positive_fn(x):
    return x>0
```

```python
list(filter(positive_fn, [3, 10, -4, 0, 5, 10, -9, 6]))
```

```python
list(filter(lambda x : x>0 , [3, 10, -4, 0, 5, 10, -9, 6]))
```

#### 파일의 입출력

```python
# 파일 입출력 : Input | Output => I/O
# 표준 입출력 함수 : input, print
# 파일 단위 입출력 함수 : open(경로,모드) w/r/a

f = open('new_text.txt', 'w')
f.close() # 자원의 사용 종료

# 파일 객체 생성시에 'w' 모드일 때 경로에 파일이 존재하지않으면 파일 생성
```

```python
# 텍스트 파일에 텍스트 데이터 쓰기
f = open('new_text.txt', 'w') # I/O 객체 생성
# 텍스트 쓰기 : write() 메서드를 사용
f.write('hello~')
```

```python
f.close()
```

```python
# 텍스트 쓰기의 특징 => 모든 데이터 갱신
f = open('new_text.txt','w')
for i in range(1, 11):
    f.write(f'{i} line \n')
f.close()
```

```python
# 파일의 데이터 읽어오기 : r
f = open('new_text.txt', 'r')
# 라인단위 읽기
print(f.read())
f.close()
```

```python
f = open('new_text.txt', 'r')
while 1:
    line = f.readline()
    if not line:
        break
    print(line, end ='')
```

```python
# 파일의 텍스트 데이터를 라인 단위의 리스트로 반환 : readlines
f = open('new_text.txt', 'r')
lines = f.readlines()

#print(lines)

for line in lines:
    print(line, end='')

f.close()
```

```python
# append : 'a' => 마지막 라인 다음 라인에 텍스트 데이터를 추가
f = open('new_text.txt', 'a')
f.write('\nnew line insert !!!')
f.close()
```

### 예외 처리

```python
# 예외 발생!
try:
    print(10/0) # 예외 발생 예상되는 구문

except Exception as e:
    print(f'에러 이유는 {e}, 프로그램을 종료합니다.')
```

```python
# try~ except~ finally
# 예외 발생 여부와 상관없이 무조건 실행하는 블록
try :
    print(10/10)
    print('종료')
except:
    print('error~!!')
finally:
    print('에러와 상관 없이 실행!!!')
```

```python
# 사칙연산을 하는 클래스 
# => 클래스 이름, 변수, 함수(메서드), 초기화 함수(생성자)
# 클래스의 이름은 대문자로 시작 => FourCal

class FourCal:
    # 클래스의 멤버 : 변수, 메서드

    # 초기화 함수 => 변수 초기화, 최초 데이터 생성
    def __init__(self, num1,num2):
        # 인스턴스 변수 생성과 데이터 할당
        self.num1 = num1
        self.num2 = num2

    # 메서드를 통해 변수(인스턴스 변수)를 생성 => 클래스를 통해서 객체로 만들면 변수가
    # 함수 생성시에 self 사용하는 이유 => 인스턴스 멤버를 사용하기 위함!
    
    def printData(self):
        print(f'num1 : {self.num1}, num2 : {self.num2}')

    def add(self):
        return self.num1+self.num2

    def mul(self):
        return self.num1*self.num2

    def div(self):
        try:
            return self.num1/self.num2
        except ZeroDivisionError:
            print("에러: 0으로 나눌 수 없습니다.")
            return 0  # 혹은 None 대신 다른 기본값을 반환하도록 설정 가능

    def sub(self):
        return self.num1-self.num2
    

if __name__ == "__main__":
    result = FourCal(0,2)
    print(result.div())
```
